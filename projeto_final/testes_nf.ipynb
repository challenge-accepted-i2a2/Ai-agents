```python
import pathlib
from google import genai
from google.genai import types
```

# Leitura e transcrição das notas fiscais em xml

```python
# Retrieve and encode the PDF byte
filepath_1 = pathlib.Path('/home/notmyname/Ai-agents/projeto_final/data/testes/NFS-e 118.2025 Diego Kempf PF.pdf')
filepath_2 = pathlib.Path('/home/notmyname/Ai-agents/projeto_final/data/testes/NFS-e 118.2025 Diego Kempf PF.pdf')
```

```python
## Transcribes first nf as xml
client = genai.Client(api_key='')

prompt = "Extraia todo o texto desta nota fiscal. Retorne apenas o texto em formato xml, sem comentários."
response = client.models.generate_content(
  model="gemini-2.5-flash",
  contents=[
      types.Part.from_bytes(
        data=filepath_1.read_bytes(),
        mime_type='application/pdf',
      ),
      prompt])
print(response.text)
```
